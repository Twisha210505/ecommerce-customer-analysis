```python
import pandas as pd
df=pd.read_csv('e-commerce.csv')
print(df.to_string())
```

```python
df.dtypes
```

```python
x=df.groupby('Pages_Viewed')['Time_Spent_on_Site_Minutes'].sum()
import matplotlib.pyplot as plt
plt.bar(x.index,x.values)
plt.xlabel("Pages Viewed")
plt.ylabel("Average Time Spent (Minutes)")
plt.title('PAGES VIEWED VS AVG TIME SPENT',fontweight='bold')
plt.show
```

```python
x = df.groupby(['Location', 'Interests']).size().unstack()
x.plot(kind='bar',stacked=True,color=["#8B0000","#B22222","#DC143C","#FF6347","#FFA07A"])
plt.xlabel('LOCATION')
plt.ylabel('NUMBER OF USERS')
plt.title("Number of Users by Location and Interest",fontweight='bold')
plt.show()
```

```python
x=df['Product_Category_Preference'].value_counts()
plt.figure(figsize=(10, 9))
plt.pie(x,labels=x.index,
        shadow=True,autopct='%1.1f%%',
        explode=[0.2,0,0,0,0],
        colors=['#66b3ff', '#ff9999', '#99ff99', '#c299ff', '#ffcc99'])
plt.title("Distribution of Users by Product Category Preference",fontweight='bold')
plt.legend(title='Product Category')
plt.show()
```

```python
x=df.groupby('Last_Login_Days_Ago')['Total_Spending'].sum()
plt.scatter(x.index,x.values,color='hotpink')
font1 = {'family':'serif','color':'black','size':9}
plt.xlabel("Recency of Last Login (Days)",fontdict=font1)
plt.ylabel("Customer Lifetime Spending",fontdict=font1)

font2 = {'family':'serif','color':'black','size':9,}
plt.title('Relationship Between User Recency and Spending',fontweight=10,fontdict=font2)
plt.tight_layout()
plt.show()
```

```python
plt.hist(df['Age'],edgecolor="#EB3DEB",color="#FFD1DC",alpha=0.8)
plt.xlabel("Age")
plt.ylabel("Number of Users")
plt.title("Age Distribution of Users", fontweight='bold')
plt.show()
```

```python
x = df['Gender'].value_counts()
plt.figure(figsize=(5,5))
plt.pie(x.values,labels=x.index,autopct='%1.1f%%',
        colors=['#66b3ff', '#ff9999', '#99ff99'],
        wedgeprops={'edgecolor': 'white','width':0.5},
        startangle=90,shadow=True)
plt.title("Gender Distribution of Users", fontweight='bold')
plt.tight_layout()
plt.show()
```

```python
x = df.groupby('Purchase_Frequency')['Total_Spending'].sum()
plt.figure(figsize=(9,5))
plt.plot(x.index, x.values, marker='o', color='#FF6347', linewidth=2)
plt.xlabel("Purchase Frequency")
plt.ylabel("Total Spending")
plt.title("Total Spending vs Purchase Frequency", fontweight='bold')
plt.grid(True, linestyle='--', alpha=0.5)
plt.show()
```
